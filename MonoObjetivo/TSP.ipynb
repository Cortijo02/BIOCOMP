## Estudio de parámetros y funciones

### Experimentos

#### Operaciones básicas

Se implementan los componentes básicos de inicialización, evaluación y selección del Algoritmo Genético aplicado al TSP. La población inicial se genera de forma aleatoria, donde cada individuo es una permutación del conjunto de ciudades, garantizando diversidad desde la primera generación. Posteriormente, la población se evalúa mediante una función de fitness y se ordena según un criterio de minimización, de manera que los mejores individuos ocupan las primeras posiciones. Sobre esta población ordenada se aplica un mecanismo de selección por torneo, en el que se eligen varios individuos al azar y se selecciona el mejor de ellos, introduciendo presión selectiva sin perder diversidad y asegurando que los individuos seleccionados sean copias independientes

```python
import random
import copy

def create (alphabet, N=100): # crea y devuelve población
    # Cada individuo es una permutación del alfabeto (p.ej. lista de ciudades)
    pop = []
    for _ in range(N):
        ind = list(alphabet)         # copia
        random.shuffle(ind)          # permutación
        pop.append(ind)
    return pop

def sort_pop(pop, fit):
    fits = [fit(ind) for ind in pop]
    idx = sorted(range(len(pop)), key=lambda i: fits[i])  # minimización
    pop_sorted = [pop[i] for i in idx]
    fits_sorted = [fits[i] for i in idx]
    return pop_sorted, fits_sorted


def select (pop, T): # devuelve un individuo seleccionado por torneo, devuelve una copia para evitar efectos laterales
    # pop se supone ya ordenada por fitness (mejor = pop[0])
    # Torneo: elige T individuos al azar y devuelve el mejor (min fitness)
    best = None
    best_pos = None
    n = len(pop)
    for _ in range(T):
        j = random.randrange(n)
        if (best_pos is None) or (j < best_pos):
            best_pos = j
            best = pop[j]
    return copy.deepcopy(best)
```

#### Cruces y mutaciones

El operador de cruce utilizado en este experimento es **Order Crossover (OX)**, un operador específicamente diseñado para problemas de optimización combinatoria con representación por permutaciones, como el TSP. A diferencia de los cruces clásicos, OX garantiza que los hijos generados sean soluciones factibles, es decir, que cada ciudad aparezca exactamente una vez en la ruta.

El **Partially Mapped Crossover (PMX)** selecciona un segmento de uno de los progenitores y lo copia directamente al descendiente, estableciendo a continuación un mapeo entre los genes intercambiados para completar el resto de la solución sin introducir duplicados. De este modo, el operador preserva información posicional de los padres y garantiza que los individuos generados sean permutaciones válidas.

```python
def ox_crossover(ind1, ind2, pcross):
    if random.random() > pcross:
        return copy.deepcopy(ind1), copy.deepcopy(ind2)

    L = len(ind1)
    a, b = sorted(random.sample(range(L), 2))

    def key(city):
        return tuple(city) if isinstance(city, list) else city

    def ox(p1, p2):
        child = [None] * L
        child[a:b+1] = copy.deepcopy(p1[a:b+1])
        used = set(key(c) for c in child[a:b+1])

        pos = (b + 1) % L
        for x in p2[b+1:] + p2[:b+1]:
            if key(x) in used:
                continue
            child[pos] = copy.deepcopy(x)
            used.add(key(x))
            pos = (pos + 1) % L

        return child

    return ox(ind1, ind2), ox(ind2, ind1)


def pmx_crossover(ind1, ind2, pcross):
    if random.random() > pcross:
        return copy.deepcopy(ind1), copy.deepcopy(ind2)

    size = len(ind1)
    start, end = sorted(random.sample(range(size), 2))

    child1 = [None] * size
    child2 = [None] * size

    child1[start:end] = copy.deepcopy(ind1[start:end])
    child2[start:end] = copy.deepcopy(ind2[start:end])

    mapping1 = {tuple(ind1[i]): tuple(ind2[i]) for i in range(start, end)}
    mapping2 = {tuple(ind2[i]): tuple(ind1[i]) for i in range(start, end)}

    def fill(child, parent, mapping):
        for i in range(size):
            if child[i] is None:
                gene = tuple(parent[i])
                while gene in mapping:
                    gene = mapping[gene]
                child[i] = list(gene)
        return child

    return fill(child1, ind2, mapping1), fill(child2, ind1, mapping2)
```

Se emplea un operador de **swap mutation** (mutate). Conceptualmente, este operador puede interpretarse como una mutación equiprobable por gen, donde cada gen tiene una probabilidad inversamente proporcional a la longitud del cromosoma ($1/L$). La mutación por inversión invierte el orden de un segmento de la ruta, la mutación por mezcla (scramble) reordena aleatoriamente las ciudades de un subsegmento y la mutación aleatoria por permutaciones selecciona de forma estocástica uno de estos operadores

```python
def mutate (ind, pmut):
    if random.random() > pmut:
        return ind

    L = len(ind)
    i, j = random.sample(range(L), 2)
    ind[i], ind[j] = ind[j], ind[i]
    return ind

def inversion_mutation(ind, pmut):
    if random.random() > pmut:
        return ind
    i, j = sorted(random.sample(range(len(ind)), 2))
    ind[i:j+1] = reversed(ind[i:j+1])
    return ind

def scramble_mutation(ind, pmut):
    if random.random() > pmut:
        return ind
    i, j = sorted(random.sample(range(len(ind)), 2))
    sub = ind[i:j+1]
    random.shuffle(sub)
    ind[i:j+1] = sub
    return ind

def random_perm_mutation(ind, pmut):
    options = [mutate, inversion_mutation, scramble_mutation]
    return random.choice(options)(ind, pmut)
```

#### Evaluación

En cada iteración, la población se evalúa y ordena según la función de fitness, se seleccionan individuos mediante torneo y se generan nuevos descendientes aplicando los operadores de cruce y mutación.

La condición de parada del algoritmo se define en términos del número máximo de evaluaciones de la función objetivo $1 M$, tal y como especifica el enunciado. Para ello, se introduce un contador explícito de evaluaciones que se incrementa cada vez que la población es evaluada

```python
import csv

def evolve(
    pop,
    fit,
    mutation,
    crossover,
    pmut,
    pcross,
    max_evals=1_000_000,
    T=2,
    elitism=False
):
    evals = 0
    gen = 0
    pop_size = len(pop)

    history = []  # (generation, min_fitness)

    while evals < max_evals:
        gen += 1

        pop_sorted, fits_sorted = sort_pop(pop, fit)
        evals += pop_size

        # Guardar mejor fitness de la generación
        history.append((gen, fits_sorted[0]))

        new_pop = []

        if elitism:
            elite = copy.deepcopy(pop_sorted[0])
            new_pop.append(elite)

        while len(new_pop) < pop_size:
            p1 = select(pop_sorted, T)
            p2 = select(pop_sorted, T)

            c1, c2 = crossover(p1, p2, pcross)

            c1 = mutation(c1, pmut)
            new_pop.append(c1)

            if len(new_pop) < pop_size:
                c2 = mutation(c2, pmut)
                new_pop.append(c2)

        pop = new_pop

        # if evals % 1_000 == 0:
        #     print(f"Evals: {evals}, Best fitness: {fits_sorted[0]}")

    pop_sorted, _ = sort_pop(pop, fit)
    return pop_sorted, history
```

- Un tamaño de torneo $T \in \{2,4,6,8,10,12\}$, que regula la presión selectiva. En conjunto, esta parametrización constituye una exploración factorial sistemática que permite analizar de forma controlada cómo interactúan los mecanismos de exploración, explotación y selección sobre el rendimiento final del algoritmo.

Se ha tomado la decisión de excluir `"elitism": False` del espacio de búsqueda ya que las pruebas iniciales sugieren que es mejor trabajar con elitismo y añadirlo a la búsqueda incrementa considerablemente el tiempo de ejecución. Hay un total de `192` combinaciones, que luego a su vez generan $192 * 10$ ejecuciones ya que cada combinación se ejecuta 10 veces. Además de que en cada ejecución tiene establecido un criterio de parada de 1 millón de evaluaciones, el tiempo se dispara.

Para poder hacer un proceso óptimo de búsqueda hemos tomado las siguientes decisiones:

- Hemos definido el problema para poder ejecutarlo usando `Loky`, tratando de sacar el máximo provecho de la CPU lanzando las 10 ejecuciones en paralelo.
- Nos guardamos los CSV de todas las combinaciones posibles con el formato esperado, para luego justificar los resultados y sacar conclusiones que veremos más adelante.
- Para poder paralelizar aún más, hemos dividido el espacio de búsqueda en 3 partes. Se han generado 3 copias del jupyter original para poder lanzar las ejecuciones en diferentes equipos, llegando a usar 30 CORES para todo el proceso.

```python
import os

def run_experiment(cfg, run_id):
    random.seed(None)
    np.random.seed(None)

    pop = create(cities)

    pop, history = evolve(
        pop,
        globaldist,
        mutation=mutations[cfg["mutation"]],
        crossover=crossovers[cfg["crossover"]],
        pmut=cfg["pmut"],
        pcross=cfg["pcross"],
        T=cfg["T"],
        max_evals=1_000_000,
        elitism=cfg["elitism"]
    )

    best_val = globaldist(pop[0])

    folder = f"Logs/{run_id}"
    os.makedirs(folder, exist_ok=True)

    csv_path = (
        f"{folder}/CodGrupo_TSP_"
        f"{cfg['mutation']}_{cfg['pcross']}_{cfg['elitism']}_{cfg['pmut']}_{cfg['crossover']}_{cfg['T']}.csv"
    )

    with open(csv_path, "w", newline="") as f:
        writer = csv.writer(f)
        writer.writerow(["generation", "min_fitness"])
        writer.writerows(history)

    return best_val, csv_path
```

```python
from joblib import Parallel, parallel_backend, delayed
import time

n_runs = 10
results = []

for cfg in configs:
    print("Running config:", cfg)

    start = time.perf_counter()

    with parallel_backend("loky", inner_max_num_threads=1):
        out = Parallel(n_jobs=10)(
            delayed(run_experiment)(cfg, i) for i in range(n_runs)
        )

    elapsed = time.perf_counter() - start

    vals = [r[0] for r in out]
    csv_files = [r[1] for r in out]

    results.append({
        "mutation": cfg["mutation"],
        "crossover": cfg["crossover"],
        "pcross": cfg["pcross"],
        "pmut": cfg["pmut"],
        "elitism": cfg["elitism"],
        "T": cfg["T"],
        "mean": np.mean(vals),
        "std": np.std(vals),
        "best": np.min(vals),
        "time_sec": elapsed,
        "files": csv_files
    })

    print(results[-1])
```

```python
# Seleccionar la mejor configuración según la media
best_cfg = min(results, key=lambda x: x["best"])

print("Mejor configuración encontrada:")
print(best_cfg)

print("\nResultados tras", n_runs * len(configs), "ejecuciones:")

print("Configuración:", best_cfg["files"])
print("Media final:", best_cfg["mean"])
print("Desviación típica:", best_cfg["std"])
print("Mejor valor:", best_cfg["best"])
```

El mejor valor obtenido es de `7.750373742666942`. Haremos un pequeño estudio para comprender el proceso.

##### Mapa de calor

```python
import seaborn as sns
import matplotlib.pyplot as plt

corr = results_df[
    ["best", "pcross", "pmut", "T"]
].corr(method="spearman")

plt.figure(figsize=(8, 6))
sns.heatmap(
    corr,
    annot=True,
    cmap="coolwarm",
    center=0,
    fmt=".2f"
)
plt.title("Spearman correlation heatmap")
plt.tight_layout()
plt.show()
```

El análisis de correlación muestra que la `pmut` es el parámetro con mayor influencia sobre la calidad final de la solución, presentando una correlación negativa con el valor óptimo. Dado que el objetivo del problema es la minimización, esta relación indica que incrementar la probabilidad de mutación tiende a producir soluciones de mejor calidad, favoreciendo la exploración del espacio de búsqueda y reduciendo el riesgo de convergencia prematura.

Por otro lado, la `pcross` no muestra una influencia significativa sobre el rendimiento, lo que sugiere que su efecto depende de interacciones con otros parámetros o del operador de cruce empleado. El `T` muestra un impacto marginal dentro del rango evaluado, indicando que incrementarlo no aporta mejoras significativas en la calidad de las soluciones.

La mayor dispersión observada en PMX indica una sensibilidad más elevada a la configuración de parámetros, pero también un mayor potencial exploratorio, permitiendo escapar de regiones subóptimas y acceder a soluciones de mayor calidad.
