```python
## CELDA AÑADIDA ##

def multiple_experiments(create, evolve, globaldist, draw, cities, num_experiments=10, pmut=0.1, pcross=0.7, ngen=10000, T=6, trace=2500, elitism=False):
    best_individual = None
    best_fitness = float('inf')
    all_best_fitness = []

    for i in range(num_experiments):
        print(f"Ejecución {i+1}:")
        pop = create(cities)
        pop = evolve(pop, globaldist, pmut=pmut, pcross=pcross, ngen=ngen, T=T, trace=trace, elitism=elitism)
        current_best_fitness = globaldist(pop[0])
        all_best_fitness.append(current_best_fitness)
        if current_best_fitness < best_fitness:
            best_fitness = current_best_fitness
            best_individual = pop[0]

    mean_fitness = np.mean(all_best_fitness)
    std_fitness = np.std(all_best_fitness)

    print("\n--- Mejor resultado global ---")
    draw(best_individual)
    print(f"Mejor fitness global: {best_fitness}")
    print(f"Media de los mejores fitness: {mean_fitness}")
    print(f"Desviación típica de los mejores fitness: {std_fitness}")
```

### Implementación funcional

```python
## CELDA AÑADIDA ##

import random
from collections import deque
import copy

'''
    Funcion que inicializa la poblacion inicial en base al alfabeto
'''
def create (alphabet, N=100): # crea y devuelve población
    pop = []
    for _ in range(N):
        individual = random.sample(alphabet, len(alphabet))  # Permutación aleatoria
        pop.append(individual)
    return pop

'''
    Funcion que, dado una poblacion y la funcion fitness, retorna una tupla 
    (poblacion de individuos ordenada, fitness de los individuos)
'''
def sort_pop (pop, fitness): 

    pop_with_fitness = [(indiv, fitness(indiv)) for indiv in pop]
    sorted_pop = sorted(pop_with_fitness, key=lambda x: x[1]) 
    return [indiv for indiv, _ in sorted_pop], [fit for _, fit in sorted_pop]


''' 
    Seleccion por torneo: Se eligen T individuos de la población (por ejemplo T=2) de
    modo equiprobable (todos los individuos tienen la misma probabilidad de ser escogidos), 
    y se añade a la nueva población el de mayor calidad.
'''
def select (pop, T): 
    tournament = random.sample(pop, T)  
    return max(tournament, key=lambda ind: fit_cities(ind)).copy()


'''
    En los cromosomas en los que no sepuede repetir un valor, como en el TSP 
    (donde una ciudad no debe aparecer dos veces en un cromosoma), es necesario aplicar operadores para 
    permutacion: se basa en elegir N posiciones fijas el padre _a_ y copiarlas al hijo _a'_, luego se rellenan
    los huecos tomando del padre _b_ los elementos que no estén en _a'_ y se añaden de modo ordenado a los huecos.
    Igual para _b'_

    Se llama Crossover con máscara de orden.

        - Param pcross: determina si se cruzan los padres para generar un hijo, o por el contrario se devuelven 
                        los padres originales
'''
def crossover (ind1, ind2, pcross): 
    if random.random() > pcross:    
        return ind1.copy(), ind2.copy()

    size = len(ind1)
    start, end = sorted(random.sample(range(size), 2))  # Obtenemos de la mascara

   
    child1 = [-1] * size
    child1[start:end] = ind1[start:end]

    child2 = [-1] * size
    child2[start:end] = ind2[start:end]

    
    def fill_child(child, parent):
        gene_nonRepeated = deque()  

        for gene in parent: 
            if gene not in child:
                gene_nonRepeated.append(gene)  # se añade al final de la cola

        for idx in range(len(child)):  
            if child[idx] == -1:  
                child[idx] = gene_nonRepeated.popleft()   

        return child

    child1 = fill_child(child1, ind2)
    child2 = fill_child(child2, ind1)

    return child1, child2

'''
    Mutacion equiprobable 1/longitud_cromosoma para cada uno de los genes de un individuo
'''
def mutate (ind, pmut): 

    for i in range(len(ind)):
        if np.random.rand() < pmut: # se escoge un indice aleatorio con el que se va a reemplazar el gen "i"
            i_2 = np.random.randint(0, len(ind))
            ind[i], ind[i_2] = ind[i_2], ind[i]
    return ind


# TODO: REVISAR la lógica de max_evals de evolve
# PDF de la práctica: Los algoritmos deben estar configurados para que su condición de parada sea 1M  evaluaciones. 

'''
    Algoritmo de evolución generacional adaptando condición de parada sea 1M evaluaciones.
'''
def evolve(pop, fit, pmut, pcross=0.7, max_evals=1_000_000, T=2, trace=0, elitism=False):

    evals = 0

    def fit_count(ind):
        nonlocal evals
        evals += 1
        return fit(ind)

    pop, fitness = sort_pop(pop, fit_count)

    generation = 0

    while evals < max_evals:
        new_poblacion = []

        if elitism:
            new_poblacion.append(pop[0].copy())

        while len(new_poblacion) < len(pop):
            parent_1 = select(pop, T)
            parent_2 = select(pop, T)

            child_1, child_2 = crossover(parent_1, parent_2, pcross)

            child_1 = mutate(child_1, pmut)
            child_2 = mutate(child_2, pmut)

            new_poblacion.extend([child_1, child_2])

        pop, fitness = sort_pop(new_poblacion[:len(pop)], fit_count)

        if trace > 0 and generation % trace == 0:
            print(f"Gen {generation} | Eval {evals} | Best {fitness[0]}")

        generation += 1

    print(f"Parada en {evals} evaluaciones | Generaciones: {generation}")
    return pop
```
